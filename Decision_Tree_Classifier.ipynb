From above it is cleared that there are 3 columns having outliers i.e.. 'Age', 'duration' and 'Campaign'. So, we will remove these outliers using Interquantile Range.

```python
# Removing outliers

column = df[['age','campaign','duration']]
q1 = np.percentile(column, 25)
q3 = np.percentile(column, 75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
df[['age','campaign','duration']] = column[(column > lower_bound) & (column < upper_bound)]
```

```python
# Plotting boxplot after removing outliers
df.plot(kind='box', subplots=True, layout=(2,5),figsize=(20,10),color='#808000')
plt.show()
```

```python
# copy with original dataframe
df1 = df.copy()
df1.columns
```

Now, the dataset is having 4119 rows and 18 columns

```python
# Conversion of categorical columns into numerical columns using label encoder.
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
df_encoded = df1.apply(lb.fit_transform)
df_encoded
```

```python
df_encoded['Term_Deposit'].value_counts() #Target variable
```

```python
x = df_encoded.drop('Term_Deposit',axis=1)  # independent variable
y = df_encoded['Term_Deposit']              # dependent variable
print(x.shape)
print(y.shape)
```

```python
print(4119*0.25)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
def eval_model(y_test,y_pred):
    acc = accuracy_score(y_test,y_pred)
    print('Accuracy_Score',acc)
    cm = confusion_matrix(y_test,y_pred)
    print('Confusion Matrix\n',cm)
    print('Classification Report\n',classification_report(y_test,y_pred))

def mscore(model):
    train_score = model.score(x_train,y_train)
    test_score = model.score(x_test,y_test)
    print('Training Score',train_score)  # Training Accuracy
    print('Testing Score',test_score)    # Testing Accuracy
```

```python
#Train the model
Model = DecisionTreeClassifier(criterion='gini',max_depth=5,min_samples_split=10)
Model.fit(x_train,y_train)
```

```python
# Evaluate the model
mscore(Model)
```

```python
# Check for prediction
ypred_dt = Model.predict(x_test)
print(ypred_dt)
```

```python
# # Evaluate the model - confusion matrix, classification Report, Accuaracy
eval_model(y_test,ypred_dt)
```

Plot Decision Tree

```python
# cn = class names, fn = feature_names
cn = ['no','yes']
fn = x_train.columns
print(fn)
print(cn)
```

```python
plot_tree(Model,feature_names=fn,class_names=cn,filled=True)
plt.show()
```

#Decision Tree 2 (using entropy criteria) for visualization

```python
# Building Decision Tree Classifier Model
dt1 = DecisionTreeClassifier(criterion='entropy',max_depth=4,min_samples_split=15)
dt1.fit(x_train,y_train)
```
